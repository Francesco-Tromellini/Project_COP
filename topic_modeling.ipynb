```python
# Imports
import ast
import nltk
import pickle
import json

import pandas as pd
import matplotlib.pyplot as plt

import gensim
from gensim import corpora
from gensim import models

import pyLDAvis
import pyLDAvis.gensim_models as gensimvis

pyLDAvis.enable_notebook()
```

```python
# Upload the data frame from the folder
df = pd.read_pickle("./df_essential_cop26.pkl")
```

```python
# Upload the data frame from the folder
df21 = pd.read_pickle("./df_essential_cop21.pkl")
```

```python
df = pd.concat([df, df21])
print(df.shape, df21.shape)
```

```python
cleaned_list = df["Speech_lemmatized"]
total_words = []

for element in cleaned_list:
    total_words += element

print(len(total_words))
```

```python
different_words = set(total_words)
different_words = list(different_words)

print(len(different_words))
# print(different_words)
```

```python
alphabetic_only = [word for word in different_words if word.isalpha() and len(word) > 4 ]
print(len(alphabetic_only))
```

```python
total_cleaned_words = []
for element in cleaned_list:  # loop over speeches in df
    # Loop through all the words in the corpus to eliminate nulls
    for word in element:
        if word not in alphabetic_only:
            element.remove(word)
            
    total_cleaned_words += element
            
print(len(total_cleaned_words))
```

```python
from collections import Counter
cnt = Counter(total_cleaned_words)
print(type(cnt))
cnt.most_common()
```

```python
sorted_words = dict(cnt.most_common())
sorted_keys = list(sorted_words.keys())
final_list = sorted_keys[21:]

print(final_list[0:50])
```

```python
for element in cleaned_list:  # loop over speeches in df
    for word in element:
        if word not in final_list or word == "change":
            element.remove(word)
       
```

```python
total_words = []

for element in cleaned_list:
    total_words += element

print(len(total_words))
print(type(cleaned_list))
```

```python
dictionary = corpora.Dictionary(cleaned_list)
corpus = [dictionary.doc2bow(text) for text in cleaned_list]

tfidf = models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]
```

```python
num_topics = 9

ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=21)
```

```python
topics = ldamodel.print_topics(num_words=30)
for topic in topics:
    print(topic)
```

```python
lda_display = gensimvis.prepare(ldamodel, corpus_tfidf, dictionary, sort_topics=False)
pyLDAvis.display(lda_display)
```
